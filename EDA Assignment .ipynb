**# Q 5. Make 5 questions and solve them by using group by.**

**1. Group by Gender with mean of the Rank**

```python
df.groupby(['Gender'])['Rank'].mean()
```

**2. Find all the values that fall under Other in FundedBys**

```python
df[df['Funded Bys'] =='Other']
```

**3. Group by Study Result with the mean of Rank**

```python
df.groupby(['Study Results'])['Rank'].mean()
```

**4. Group by Gender with the aggregate of Age Count and Study Result nunique**

```python
df.groupby(['Gender']).agg({'Age':'count','Study Results':'nunique'})
```

**5. Group by Study Type with the mean of Enrollment**

**# Q 6. Make 5 questions and solve them by using conditional statements.**

**1. Find Active, not recruiting and Not yet recruiting observations fall under Status**

```python
df.loc[df.Status.isin(['Active, not recruiting', 'Not yet recruiting'])]
```

**2. Find Covid19 and COVID observaions fall under Conditions**

```python
df.loc[df.Conditions.isin(['Covid19','COVID'])]
```

**3. Find all observations that fall under Status as "Suspended" and Gender as "All"**

```python
df.loc[(df.Gender == 'All') & (df.Status == 'Suspended')]
```

**4. Find "Interventional" & "Observatinal" observations that come under Study Type**

```python
df.loc[(df['Study Type'] == 'Interventional') | (df['Study Type'] =='Observational')]
```

**5. Find all the values with status as 'Withdrawn' and Gender as 'All'**

```python
df.loc[(df.Status == 'Withdrawn') | (df.Gender == 'All')]
```
